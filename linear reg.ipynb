```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Fit a simple LR: Study Hours → Exam Score

```python
df=pd.read_csv("study_hours_scores_150.csv")
```

```python
model=LinearRegression()
model.fit(df[["hours"]],df["score"])
y_pred=model.predict(df[["hours"]])
plt.scatter(df["hours"],df["score"])
plt.plot(df["hours"],y_pred,color="red")
plt.show()
```

```python
print(model.coef_[0])
print(model.intercept_)
print(model.score(df[["hours"]],df["score"]))
```

```python
pred_value=model.predict([[9]])
print(pred_value[0])
```

Fit LR: House Size → Price

```python
a=pd.read_csv("house_price_size_150.csv")
model_2=LinearRegression()
model_2.fit(a[["size_m2"]],a["price"])
y2_pred=model_2.predict(a[["size_m2"]])
plt.scatter(a["size_m2"],a["price"])
plt.plot(a["size_m2"],y2_pred,color="red")
plt.show()
```

```python
pred_2=model_2.predict([[120]])
print(pred_2[0])
```

Salary vs Experience

```python
b=pd.read_csv("salary_experience_150.csv")
model_3=LinearRegression()
model_3.fit(b[["experience_years"]],b["salary_k"])
y3_pred=model_3.predict(b[["experience_years"]])
plt.scatter(b["experience_years"],b["salary_k"])
plt.plot(b["experience_years"],y3_pred,color="red")
plt.show()
```

```python
pred_3=model_3.predict([[7]])[0]
print(pred_3)
```

Multiple Linear Regression: House Price ~ Size + Bedrooms

```python
c=pd.read_csv("house_price_multifeature_150.csv")
model_4=LinearRegression()
model_4.fit(c[["size_m2","bedrooms"]],c["price_k"])
print(model_4.coef_[0])
print(model_4.intercept_)
print(model_4.score(c[["size_m2","bedrooms"]],c["price_k"]))
pred_4=model_4.predict([[85,3]])[0]
print(pred_4)
```

Multiple Features: Advertising Spend → Sales

```python
d=pd.read_csv("advertising_linear_150.csv")
model_5=LinearRegression()
model_5.fit(d[["TV_spend","radio_spend","social_spend"]],d["sales_k"])
print(model_5.coef_[0])
print(model_5.intercept_)
print(model_5.score(d[["TV_spend","radio_spend","social_spend"]],d["sales_k"]))
pred_5=model_5.predict([[160,32,18]])[0]
print(pred_5)
```
